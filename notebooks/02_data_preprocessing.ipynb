# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Loading Dataset

```python
df = pd.read_csv("/Users/mahdisahil/Desktop/Predicting-LIfe-Expectancy/data/raw/Life Expectancy Data.csv")
```

# Clean Columns Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(r"\s+", " ", regex=True)
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace("/", "_")
)

df.columns.to_list()
```

# Remove duplicates

```python
print(f"before = {df.duplicated().sum()}")
df.drop_duplicates(inplace=True)
print(f"after = {df.duplicated().sum()}")
```
